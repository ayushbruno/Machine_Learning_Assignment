```python
sns.countplot(x='LUNG_CANCER',data=df)
plt.title('Distribution of Lung Cancer Cases')
plt.xlabel('Lung Cancer')
plt.ylabel('Count')
plt.show()
```

```python
sns.barplot(x='AGE',y='LUNG_CANCER',data=df)
plt.title('Age Distribution by Lung Cancer')
plt.xlabel('Lung Cancer')
plt.ylabel('Age')
plt.show()
```

```python
sns.countplot(data=df,x='GENDER',hue='LUNG_CANCER')
plt.title('Gender Distribution by Lung Cancer')
plt.xlabel('Gender')
```

```python
sns.countplot(data=df,x='SMOKING',hue='LUNG_CANCER')
plt.title('Smoking Distribution by Lung Cancer')
```

```python
df['GENDER']= df['GENDER'].replace({'M':1,'F':0})
```

```python
df['LUNG CANCER']= df['LUNG_CANCER'].replace({'YES':1,'NO':0})
```

```python
df['GENDER'].value_counts()
```

```python
df['LUNG CANCER'].value_counts()
```

```python
df.info()
```

```python
if df['LUNG_CANCER'].dtype == 'object':
    df = df.drop(columns=['LUNG_CANCER'])
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
sns.histplot(data=df,x='AGE',hue="LUNG CANCER")
plt.title('Age Distribution by Lung Cancer')
plt.xlabel('Age')
plt.show()
# 0 indicates NO and 1 indicates YES
```

```python
sns.boxplot(data=df,x='SMOKING',y='AGE')
plt.title('Smoking Distribution by Age')
plt.show()
```
